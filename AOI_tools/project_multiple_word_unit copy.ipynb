```python
from subAOI import SubAOI
import json
import pandas as pd
```

```python
aoi_by_word = SubAOI(fontname = "NetflixSans-Bold.otf",
            fontsize_in_px=44,
            subtitlepath = "Ep1DUTCHCOMPLETEeyetracking.srt",
            screensize =(1920,1080),
            vertical_offset=0.8163,
            vertical_offset_top=0.1466)
```
