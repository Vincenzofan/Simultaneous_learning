```python
from subAOI import SubAOI
import json
import pandas as pd
```

```python
aoi_by_word = SubAOI(fontname = "NetflixSans-Bold.otf",
            fontsize_in_px=44,
            subtitlepath = "Ep1DUTCHCOMPLETEeyetracking.srt",
            screensize =(1920,1080),
            vertical_offset=0.8163,
            vertical_offset_top=0.1466)
```

```python
results = aoi_by_word.calculate(
    line_spacing=10, first_line = (25, 17, 25, 5), second_line = (25, 5, 25, 17), single_line = (25, 11, 25, 11)
)
#1.5x padding: first_line = (25, 17, 25, 5), second_line = (25, 5, 25, 17), single_line = (25, 11, 25, 11)
#1.75x padding: first_line = (25, 28, 25, 5), second_line = (25, 5, 25, 28), single_line = (25, 17, 25, 17)
```

```python
with open("target_MWU.json") as input_targets:
    targets = json.load(input_targets)
    
aoi_MWU = SubAOI(fontname="NetflixSans-Bold.otf",
                    fontsize_in_px=44,
                    subtitlepath="Ep1DUTCHCOMPLETEeyetracking.srt",
                    screensize=(1920, 1080),
                    vertical_offset=0.8163,
                    vertical_offset_top=0.1466,
                    target_words=targets)
results_MWU = aoi_MWU.calculate(
    line_spacing=10, first_line = (0, 17, 0, 5), second_line = (0, 5, 0, 17), single_line = (0, 11, 0, 11) # no padding for MWU
)
```

```python
aoi_set_1_50 = pd.concat([results,results_MWU])
```
